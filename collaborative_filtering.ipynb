```python
import numpy as np
import csv
import time
```

```python
train_file = 'train.csv'
test_file  = 'test.csv'
soln_file  = 'user_median.csv'
profiles_file = 'profiles.csv'
artists_file = 'artists.csv'
```

```python
# Load list of users
users = []
with open(profiles_file, 'r') as profiles_fh:
    profiles_csv = csv.reader(profiles_fh, delimiter=',', quotechar='"')
    next(profiles_csv, None)
    for row in profiles_csv:
        user = row[0]
        users.append(user)
```

```python
# Load list of artists
artists = []
with open(artists_file, 'r') as profiles_fh:
    artists_csv = csv.reader(profiles_fh, delimiter=',', quotechar='"')
    next(artists_csv, None)
    for row in artists_csv:
        artist = row[0]
        artists.append(artist)
```

```python
artist_sims = np.zeros([len(artists), len(artists)])
```

```python
# Load the training data.
train_data = {}
with open(train_file, 'r') as train_fh:
    train_csv = csv.reader(train_fh, delimiter=',', quotechar='"')
    next(train_csv, None)
    for row in train_csv:
        user   = row[0]
        artist = row[1]
        plays  = row[2]
        if not user in train_data:
            train_data[user] = {}
        train_data[user][artist] = int(plays)
```

```python
len(train_data)
```

```python
def get_user_plays(user, artist):
    if user in train_data:
        for a, p in train_data[user].iteritems():
            if artist == a:
                return p
    return 0.0
```

```python
# Compute the global median.
plays_array = []
for user, user_data in train_data.iteritems():
    for artist, plays in user_data.iteritems():
        plays_array.append(plays)
global_median = np.median(np.array(plays_array))
print "global median:", global_median
```

```python
# Code to calculate user biases
user_dict = {}
for user, user_data in train_data.iteritems():
    plays_list = []
    for artist, plays in user_data.iteritems():
        plays_list.append(plays)
    plays_average = np.average(plays_list)
    user_dict[user] = plays_average
```

```python
line = 0
# Write out test solutions.
with open(test_file, 'r') as test_fh:
    test_csv = csv.reader(test_fh, delimiter=',', quotechar='"')
    next(test_csv, None)

    with open(soln_file, 'w') as soln_fh:
        soln_mean_csv = csv.writer(soln_fh,
                              delimiter=',',
                              quotechar='"',
                              quoting=csv.QUOTE_MINIMAL)
        soln_mean_csv.writerow(['Id', 'plays'])

        for row in test_csv:
            line += 1
            if line % 1000 == 0:
                print line, row

            id     = row[0]
            user   = row[1]
            artist = row[2]

            if (user_dict[user]-global_median) < 50:
                soln_mean_csv.writerow([id, user_dict[user]])
                
            else:
                soln_mean_csv.writerow([id, global_median])
```

```python
# Calculate the similarities between two artists. 
# This is done by: 

for i in range(len(artist)):
    for j in range(len(artist)):
        if i == j: 
            artist_sims[i][j] = global_median
        else:
            for train_user, user_data in train_data.iteritems():  
                artist_i_plays = 0
                count_i = 0 
                artist_j_plays = 0
                count_j = 0
                for train_artist, plays in user_data.iteritems():
                    if (train_artist == artists[i]):
                        artist_i_plays = plays
                        count_i+=1
                
                    if (train_artist == artists[j]):
                        artist_j_plays+=plays
                        count_j+=1


                    if ((count_i + count_j) > 1):
                        mean_i_j = ((artist_i_plays/count_i) + (artist_j_plays/count_j))/2        
                        artist_sims[i][j] = mean_i_j

                    
                    
                
```
